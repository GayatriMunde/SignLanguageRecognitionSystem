```python
import numpy as np
import cv2
import keras
import tensorflow as tf
```

```python
model = keras.models.load_model("sample_model.h5")
cam = cv2.VideoCapture(0)
number_dict = {0:'Zero',1:'One',2:'Two',3:'Three',4:'Four',5:'Five',6:'Six',7:'Seven',8:'Eight',9:'Nine',10:'A',11:'B',12:'C',13:'D',14:'E',15:
              'F',16:'G',17:'H',18:'I',19:'J',20:'K',21:'L',22:'M',23:'N',24:'O',25:'P',26:'Q',27:'R',28:'S',29:'T',30:'U',31:'V',32:'W',33:'X',34:'Y',35:'Z'}
```

```python
while True:
    _, frame = cam.read()
    frame = cv2.flip(frame, 1)
    #cv2.rectangle(frame, (319, 9), (620 + 1, 309), (0, 255, 0), 1)
    #roi = frame[10:300, 320:620]

    # cv2.imshow("Frame", frame)
    
    
    x1 = 320
    y1 = 10
    x2 = 630
    y2 = 320
    # Drawing the ROI
   
    cv2.rectangle(frame, (x1-1, y1-1), (x2+1, y2+1), (255,0,0) ,1)
    # Extracting the ROI
    roi = frame[y1:y2, x1:x2]
    roi = cv2.resize(roi, (256, 256)) 
    
    
    gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
    gaussblur = cv2.GaussianBlur(gray, (5, 5), 2)
    smallthres = cv2.adaptiveThreshold(gaussblur, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 11, 2)
    ret, final_image = cv2.threshold(smallthres, 70, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
    cv2.imshow("BW", final_image)
    final_image = cv2.resize(final_image, (256, 256))

    final_image = np.reshape(final_image, (1, final_image.shape[0], final_image.shape[1], 1))
    pred = model.predict(final_image)
    print(number_dict[np.argmax(pred)])
    cv2.putText(frame,number_dict[np.argmax(pred)], (10, 50), cv2.FONT_HERSHEY_PLAIN, 1, (0, 255, 0), 1)
    cv2.imshow("Frame", frame)
    k = cv2.waitKey(1) & 0xFF
    if k == 27:
        break

# Release the camera and destroy all the windows
cam.release()
cv2.destroyAllWindows()
```
